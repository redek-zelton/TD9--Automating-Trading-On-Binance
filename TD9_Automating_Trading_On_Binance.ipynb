## Get all cryptocurrencies in the Binance Market

```python
# Get a list of all available cryptocurrencies and display it

def getListCrypto():
    List=[]
    resp = requests.get('https://api.binance.com/api/v1/exchangeInfo')
    List_data = resp.json()
    for i in List_data['symbols']:
        List.append(i['baseAsset'])
        List = list(dict.fromkeys(List)) # drop duplicate
    return List
```

```python
List_crypto = getListCrypto()
print('Total Crypto in Binance: ',len(List_crypto))
print(List_crypto)
```

## Get the order book from a pair 

```python
def getDepth(direction='asks',pair='BTCUSDT'):
    resp = requests.get('https://api.binance.com/api/v3/depth?symbol='+pair)
    Liste_data = resp.json()
    List = Liste_data[direction]
    return List
```

## Get Candles

```python
def refreshDataCandle(pair='BTCUSDT',duration='5m'):
    List=[]
    Liste_attribut=['Open time','Open','High','Low','Close','Volume','Close time','Quote asset volume'
                   ,'Number of trades','Taker buy base asset volume','Taker buy quote asset volume'
                   ,'Ignore']
    
    #example schema per candle
    """
    [
    1499040000000,      // Open time
    "0.01634790",       // Open
    "0.80000000",       // High
    "0.01575800",       // Low
    "0.01577100",       // Close
    "148976.11427815",  // Volume
    1499644799999,      // Close time
    "2434.19055334",    // Quote asset volume
    # Below, not need
    308,                // Number of trades
    "1756.87402397",    // Taker buy base asset volume
    "28.46694368",      // Taker buy quote asset volume
    "17928899.62484339" // Ignore
    ]
    """
    resp = requests.get('https://api.binance.com/api/v3/klines?symbol='+pair+'&interval='+duration)
    Liste_data = resp.json()
    for data in Liste_data:
        data=data[:8]
        data=[int(data[0]),float(data[1]),float(data[2]),float(data[3]),float(data[4]),float(data[5]),int(data[6]),float(data[7])]
        P=zip(Liste_attribut,data)
        P=dict(P)
        List.append(P)
    return List
```

```python
pair='BTCUSDT'
duration='5m'
refreshDataCandle(pair=pair,duration=duration)
```

# Use SQLITE

```python
# Creér une Table SQL avec la Pair
# Create a sqlite table to store said data 
def createTable(pair):
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
    cursor.execute(
        '''
        CREATE TABLE '''+pair+
        '''
        (
        Id INTEGER PRIMARY KEY, 
        date INT, 
        high REAL, 
        low REAL, 
        open REAL,
        close REAL, 
        volume REAL, 
        quotevolume REAL
        )
        '''
    )
    connexion.close()
    return True
```

```python
# Avoir l'Index de la dernière mise à jour
def getindex(pair):
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
    Id=0
    sql ="SELECT Max(Id) FROM "+pair
    Ide = cursor.execute(sql)
    for index in Ide:
        Id=index
    connexion.close()
    return Id
```

```python
# Avoir la dernière date de Mise à Jour
def getLastDate(pair):
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
    Date=0
    sql ="SELECT Max(date) FROM "+pair
    Ide = cursor.execute(sql)
    for index in Ide:
        Date=index
    connexion.close()
    return Date
```

```python
# Utilise la dernière date de Mise à Jour pour enlever les données expirées
def getAvailableData(pair, List):
    cpt=0
    time = getLastDate(pair)[0]
    if time == None:
        time = 0
    for data in List:
        if time < data['Close time']:
            return List[cpt:]
        else:
            cpt+=1
```

```python
# Mettre les Modifications
# Store candlModify function to update when new candle data is available in the db
def StorecandlModify(pair,List):
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
    
    # Get Index
    index=getindex(pair)[0]
    if index == None:
        # Nouvelle Table
        index=0
    else:
        index+=1
    
    # Available Data
    List = getAvailableData(pair, List)
    if List==None:
        return False
        
    # Insertion
    for data in List:
        sql = "INSERT INTO "+pair+" (Id,date,high,low,open,close,volume,quotevolume) VALUES (?,?,?,?,?,?,?,?)"
        value =(index,data['Close time'],data['High'],data['Low'],data['Open'],data['Close'],data['Volume'],data['Quote asset volume'])
        cursor.execute(sql,value)
        connexion.commit()
        index+=1
    connexion.close()
    return True
```

```python
# Create new Table with Update
pair='BTCUSDT'
print('Créer Table => ',createTable(pair))
print('Last Index => ',getindex(pair)[0])
print('Last Data => ',getLastDate(pair)[0])
List =refreshDataCandle(pair,'5m')
print('Update => ',StorecandlModify(pair,List))
```

```python
# Only Update
List =refreshDataCandle(pair,'5m')
print('Update => ',StorecandlModify(pair,List))
print('Last Index => ',getindex(pair)[0])
print('Last Data => ',getLastDate(pair)[0])
```

## Create an Order

```python
# Add Order in database
def AddOrder(uuid,symbol,quantity,price,timestamp,side):
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
        
    # Insertion
    sql = "INSERT INTO OrderBook (uuid,symbol,quantity,price,timestamp,side) VALUES (?,?,?,?,?,?)"
    value =(uuid,symbol,quantity,price,timestamp,side)
    cursor.execute(sql,value)
    connexion.commit()
    connexion.close()
    return True
```

```python
def createOrder(api_key="",secret_key="",direction='SELL',price=0,amount=0,pair='XTZUSDT',orderType='LIMIT'):
    client = Client(api_key, secret_key)
    order = client.create_order(
    symbol=pair,
    side=direction,
    type=orderType,
    quantity=amount,
    price=price,
    timeInForce=client.TIME_IN_FORCE_GTC)
    
    #Add database OrderBook
    AddOrder(order['orderId'],order['symbol'],float(order['origQty']),float(order['price']),int(order['transactTime']),order['side'])
    return order
```

## Cancel an Order

```python
def findOrder(uuid):
    List=[]
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
    sql = "SELECT uuid,symbol FROM OrderBook WHERE uuid="+uuid
    Liste_data = cursor.execute(sql)
    for data in Liste_data:
        List.append(data)
    connexion.close()
    return List[0]
```

```python
def delOrder(uuid):
    connexion = sqlite3.connect('Test.db')
    cursor = connexion.cursor()
    sql = "DELETE FROM OrderBook WHERE uuid="+uuid
    delete = cursor.execute(sql)
    connexion.commit()
    return True
```

```python
def cancelOrder(api_key, secret_key, uuid):
    client = Client(api_key, secret_key)
    order=findOrder(uuid)
    
    result = client.cancel_order(
    symbol=order[1],
    orderId=order[0])
    
    delOrder(uuid)
    return result
```
